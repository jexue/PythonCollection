```python
col = 1
while col < len(df.columns):
    if len(df.iloc[:,col].unique()) == 2:
        df.iloc[:,col] = pd.factorize(df.iloc[:,col])[0]
        col = col + 1
    else:
        df.iloc[:,col] = (pd.factorize(df.iloc[:,col])[0] + 1) / len(df.iloc[:,col].unique()) 
        col = col + 1

print(df)
```

```python
# Delete columns

del df['ItemCreateDate(dd-mmm-yyyy)']
del df['Error1']
del df['Error2']
```

```python
df_array = df.iloc[:, 1:len(df)].values
sample = numpy.random.rand(len(df_array)) < 0.8

train = df_array[sample]
test = df_array[~sample]

var = len(df.columns) - 2

X = train[:,0:var]
Y = train[:,var]
          
print(Y)
```

```python
model = Sequential()
model.add(Dense(var,input_dim=var,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
print(model.input_shape)
```

```python
model.compile(loss='binary_crossentropy',  optimizer='adam', metrics=['accuracy'])
model.fit(X,Y,epochs=150, batch_size=10)
```

```python
#Create Test Set
X_Test = test[:,0:var]
Y_Test = test[:,var]

#evaluate model using test set data
scores = model.evaluate(X_Test,Y_Test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
